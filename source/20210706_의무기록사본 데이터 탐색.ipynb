# 신장 의무기록사본

### 1. 라이브러리 로드

```python
import pandas as pd
import numpy as np
from glob import glob
```

### 2. 데이터 로드

```python
data = pd.read_excel('D:/05.Kidney/data/개별데이터/2013-03-11.xlsx')
column_name = list(data.columns)
df = pd.DataFrame(columns= column_name)

data_lst = glob('D:/05.Kidney/data/개별데이터/*.xlsx')

for i in data_lst:
    #print(i)
    dat = pd.read_excel(i)
    df = pd.concat([df, dat])
```
